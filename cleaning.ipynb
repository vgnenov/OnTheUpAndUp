```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import datetime as dt
```

```python
csvpath = "Illinois_Mean_Salary.csv"
mean_salary = pd.read_csv(csvpath)
mean_salary.head(10)
```

```python
nan_value = float("NaN")
mean_salary.replace("", nan_value, inplace = False)
mean_salary.head(10)
```

```python
reduced_mean_salary = mean_salary [["Year", "Mean Hourly Wage", "Annual Mean Wage", "Median Hourly Wage", "Median Weekly Earnings", "Annual Median Wage"]]
reduced_mean_salary.head(10)
```

```python
reduced_mean_salary = reduced_mean_salary.dropna(how='any')
reduced_mean_salary.head(20)
```

```python
reduced_mean_salary.to_csv("Cleaned_Illinois_Mean_Salary.csv", encoding="utf-8", index=False)
```
