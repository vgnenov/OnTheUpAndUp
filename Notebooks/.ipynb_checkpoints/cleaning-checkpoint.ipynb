```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import datetime as dt
```

```python
csvpath = "Copy of all_data_M_2019.csv"
employment_df = pd.read_csv(csvpath)
employment_df.head(10)
```

```python
nan_value = float("NaN")
employment_df.replace("", nan_value, inplace = False)
employment_df.head(10)
```

```python
reduced_df = employment_df [["area_title", "naics", "own_code", "occ_code", "h_median", "h_pct75", "h_pct90", "a_pct10", "a_pct25", "a_median", "a_pct75", "a_pct90"]]
reduced_df.head(10)
```

```python
reduced_df = reduced_df.dropna(how='any')
reduced_df.head(20)
```

```python
reduced_df.to_csv("Copy_of_all_data_M_2019_cleaned.csv", encoding="utf-8", index=False)
```
